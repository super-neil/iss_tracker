```python
# Imports & Setup
import numpy as np
import matplotlib.pyplot as plt
from skyfield.api import load, wgs84
from skyfield import almanac

# Enable inline plotting
%matplotlib inline
plt.style.use('dark_background')

# 1. Define Observer Location (Banchory, Scotland)
lat_degrees = 57.0519
lon_degrees = -2.4839
my_location = wgs84.latlon(lat_degrees, lon_degrees)

# 2. Load Astronomical Data
print("Loading planetary data...")
eph = load('de421.bsp')
sun = eph['sun']
earth = eph['earth']

# 3. Load Satellite Data (TLE)
print("Downloading latest ISS TLE data...")
stations_url = 'http://celestrak.org/NORAD/elements/stations.txt'
satellites = load.tle_file(stations_url)

# Convert the list of satellites into a dictionary indexed by name
by_name = {sat.name: sat for sat in satellites}
iss = by_name['ISS (ZARYA)']

ts = load.timescale()

print(f"Setup Complete. Observer set to {lat_degrees}Â°N, {lon_degrees}Â°W")
```

```python
# Real-Time Tracker

# Get current time
t_now = ts.now()

# Calculate Geocentric Position (Space)
geocentric = iss.at(t_now)

# Calculate Subpoint (Ground)
subpoint = wgs84.subpoint(geocentric)

# Calculate Position Relative to location (Topocentric)
difference = iss - my_location
topocentric = difference.at(t_now)
alt, az, distance = topocentric.altaz()

print(f"â±ï¸  TIME: {t_now.utc_strftime('%Y-%m-%d %H:%M:%S')} UTC")
print("-" * 40)
print(f"ðŸŒ ISS GROUND LOCATION:")
print(f"   Latitude:  {subpoint.latitude.degrees:.4f}")
print(f"   Longitude: {subpoint.longitude.degrees:.4f}")
print(f"   Altitude:  {subpoint.elevation.km:.1f} km")
print("-" * 40)
print(f"ðŸ”­ FROM BANCHORY (Scotland):")
if alt.degrees > 0:
    print(f"   STATUS:  ABOVE HORIZON! (Look {az.degrees:.0f}Â°)")
else:
    print(f"   STATUS:  Below Horizon")
    
print(f"   Elevation: {alt.degrees:.2f}Â°")
print(f"   Distance:  {distance.km:.0f} km")
```

```python
# Find Next Visible Pass

def find_next_visible_pass(days_to_scan=30):
    t0 = ts.now()
    t1 = ts.now() + days_to_scan
    
    print(f"Scanning the next {days_to_scan} days for visible passes...")
    
    # 1. Find all geometric passes (ISS above 10 deg)
    # This might return a lot of data for 30 days, so we process carefully
    times, events = iss.find_events(my_location, t0, t1, altitude_degrees=10.0)
    
    visible_passes = []
    
    # Iterate through events in groups of 3 (Rise -> Culminate -> Set)
    for i in range(0, len(events), 3):
        try:
            # We check the "Culmination" (Peak) time for visibility
            t_rise = times[i]
            t_culm = times[i+1]
            t_set = times[i+2]
            
            # Check Sun Angle (Must be < -6 deg for Civil Twilight/Darkness)
            observer = earth + my_location
            sun_alt = observer.at(t_culm).observe(sun).apparent().altaz()[0].degrees
            
            # Check ISS Lighting (Must be sunlit)
            iss_sunlit = iss.at(t_culm).is_sunlit(eph)
            
            if sun_alt < -6 and iss_sunlit:
                visible_passes.append({
                    'rise': t_rise,
                    'culminate': t_culm,
                    'set': t_set
                })
                
        except IndexError:
            pass # Handle partial pass at end of data

    return visible_passes

# Run the search
passes = find_next_visible_pass(days_to_scan=30)

if passes:
    next_pass = passes[0] # The very next one
    print(f"ðŸŽ‰ FOUND {len(passes)} VISIBLE PASSES in the next 30 days!")
    print(f"Next Sighting: {next_pass['rise'].utc_strftime('%Y-%m-%d %H:%M:%S')} UTC")
    
    # Optional: Print the next 3 just so you have options
    print("\nTop 3 Upcoming Opportunities:")
    for p in passes[:3]:
        print(f" - {p['rise'].utc_strftime('%a %d %b %H:%M')} UTC")
else:
    print("No visible passes found even in the next 30 days. (The orbit might be in a daylight cycle).")
```

```python
# Sky Chart Visualization

if not passes:
    print("Cannot plot: No visible passes found.")
else:
    # Use the next pass found in the previous cell
    p_start = passes[0]['rise']
    p_end = passes[0]['set']
    
    # 1. Generate High-Res Data for the Plot
    # Create a time point for every second of the pass
    duration_mins = (p_end - p_start) * 24 * 60
    num_seconds = int(duration_mins * 60)
    time_vector = p_start + np.linspace(0, duration_mins, num_seconds) / (24*60)
    
    # Calculate Alt/Az for the whole path
    difference = iss - my_location
    topocentric = difference.at(time_vector)
    alt, az, _ = topocentric.altaz()
    
    # 2. Setup Polar Plot
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='polar')
    
    # Rotate to put North on top
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1) # Clockwise
    
    # Map Altitude (0-90) to Radius (90-0)
    # Center = Zenith (90 deg alt), Edge = Horizon (0 deg alt)
    r = 90 - alt.degrees
    theta = az.radians
    
    # 3. Plot the Path
    # Mask out any data below horizon (just in case)
    mask = r < 90
    ax.plot(theta[mask], r[mask], color='cyan', linewidth=3, alpha=0.8, label='ISS Trajectory')
    
    # Mark Start (Green Dot)
    ax.plot(theta[mask][0], r[mask][0], 'o', color='lime', markersize=12, label='Rise')
    
    # Mark End (Red X)
    ax.plot(theta[mask][-1], r[mask][-1], 'x', color='red', markersize=12, label='Set')
    
    # 4. Styling
    ax.set_rmax(90)
    ax.set_yticks([0, 30, 60, 90])
    ax.set_yticklabels(['Zenith', '60Â°', '30Â°', 'Horizon'], color='white')
    
    # Title & Legend
    title_str = f"Next ISS Sighting: {p_start.utc_strftime('%a %d %b %H:%M')} UTC"
    ax.set_title(title_str, color='white', fontsize=14, pad=20)
    
    # Dark Theme Tweak
    ax.set_facecolor('#050515')
    fig.patch.set_facecolor('#000005')
    ax.grid(True, color='gray', alpha=0.3)
    ax.tick_params(colors='gray')
    
    # Custom Legend
    legend = ax.legend(loc='lower left', bbox_to_anchor=(1.05, 0.1), facecolor='#050515')
    for text in legend.get_texts():
        text.set_color("white")
        
    plt.show()
```

```python
# Ground Track
import matplotlib.patches as mpatches

if not passes:
    print("No pass to map.")
else:
    # Reuse the time vector from Cell 4
    # Calculate Geocentric (Lat/Lon) for the path
    geocentric_path = iss.at(time_vector)
    subpoint_path = wgs84.subpoint(geocentric_path)
    
    lats = subpoint_path.latitude.degrees
    lons = subpoint_path.longitude.degrees
    
    # Plot
    fig, ax = plt.subplots(figsize=(12, 6))
    
    # Draw simple map background
    ax.imshow(np.zeros((180, 360)), extent=[-180, 180, -90, 90], cmap='Greys_r')
    ax.grid(True, color='gray', linestyle=':', alpha=0.5)
    
    # Plot Path
    ax.plot(lons, lats, color='cyan', linewidth=2)
    
    # Plot Observer (You)
    ax.plot(lon_degrees, lat_degrees, 'r*', markersize=15, label='Banchory')
    
    # Plot Start/End
    ax.plot(lons[0], lats[0], 'go', label='Rise')
    ax.plot(lons[-1], lats[-1], 'rx', label='Set')
    
    ax.set_xlabel("Longitude")
    ax.set_ylabel("Latitude")
    ax.set_title("Ground Track of Next Pass")
    ax.legend()
    
    plt.show()
```
